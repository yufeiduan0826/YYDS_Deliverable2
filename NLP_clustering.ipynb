```python
from gensim.models import KeyedVectors

model = KeyedVectors.load_word2vec_format(word2vec_output_file, binary=False)
```

```python
import re

def preprocess(text):
    text = str(text).lower()
    return re.findall(r'\b[a-z]+\b', text)  # keep only alphabetic tokens

df['tokens'] = df['Description'].apply(preprocess)
```

```python
import numpy as np

def get_avg_vector(tokens, model):
    vectors = [model[word] for word in tokens if word in model]
    if not vectors:
        return np.zeros(model.vector_size)
    return np.mean(vectors, axis=0)

df['vector'] = df['tokens'].apply(lambda x: get_avg_vector(x, model))

# Stack into matrix for clustering
X = np.vstack(df['vector'].values)
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertias = []
K_range = range(1, 11)  # Try k from 1 to 10

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertias.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8, 5))
plt.plot(K_range, inertias, marker='o')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia (within-cluster sum of squares)')
plt.title('Elbow Method to Determine Optimal k')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=8, random_state=42)  # Change n_clusters as needed
df['cluster'] = kmeans.fit_predict(X)
```

```python
for i in sorted(df['cluster'].unique()):
    print(f"\nCluster {i}:")
    print(df[df['cluster'] == i]['Description'].head(5).tolist())
```

```python
for cluster_id in sorted(df['cluster'].unique()):
    unique_descriptions = df[df['cluster'] == cluster_id]['Description'].unique()
    print(f"\nCluster {cluster_id} - {len(unique_descriptions)} unique descriptions:")
    for desc in unique_descriptions[:10]:  # change [:10] to see more
        print(f"  - {desc}")
```

```python
df['joined'] = df['tokens'].apply(lambda x: ' '.join(x))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

top_n = 10  # Number of keywords per cluster

for cluster_id in sorted(df['cluster'].unique()):
    cluster_docs = df[df['cluster'] == cluster_id]['joined']

    vectorizer = TfidfVectorizer()
    tfidf_matrix = vectorizer.fit_transform(cluster_docs)
    feature_names = vectorizer.get_feature_names_out()

    # Get mean TF-IDF score for each term in the cluster
    mean_tfidf = tfidf_matrix.mean(axis=0).A1
    top_indices = mean_tfidf.argsort()[::-1][:top_n]
    top_keywords = [feature_names[i] for i in top_indices]

    print(f"\n🔹 Cluster {cluster_id} Top {top_n} Keywords:")
    print(', '.join(top_keywords))
```
